```python
import cv2
import numpy as np
import os
import random

# ------------------------------
# PARAMETRI DI BASE
# ------------------------------
img_size = 128           # dimensione immagini (128x128 pixel)
num_train = 350
num_test = 150

output_dir = "Dataset"
test_dir = "Testing"

# Crea cartelle per le due classi
os.makedirs(f"{output_dir}/circles", exist_ok=True)
os.makedirs(f"{output_dir}/squares", exist_ok=True)
os.makedirs(f"{test_dir}", exist_ok=True)
```

```python
# ------------------------------
# FUNZIONE PER GENERARE IMMAGINI
# ------------------------------

def genera_cerchio():
    img = np.ones((img_size, img_size), dtype=np.uint8) * 255  
    center = (random.randint(20, 108), random.randint(20, 108))
    radius = random.randint(10, 50)
    cv2.circle(img, center, radius, 0, -1)  
    return img

def genera_quadrato():
    
    img = np.ones((img_size, img_size), dtype=np.uint8) * 255  
    x1 = random.randint(10, 90)
    side = random.randint(20, 40)
    cv2.rectangle(img, (x1, x1), (x1 + side, x1 + side), 0, -1)  
    
    
    if random.random() < 0.5:  # 50% di probabilità di rotazione
        angle = random.uniform(0, 90)  # angolo casuale tra 0 e 90 gradi
        # Centro dell'immagine
        center = (img_size // 2, img_size // 2)
        # Matrice di rotazione
        rot_mat = cv2.getRotationMatrix2D(center, angle, 1.0)
        # Ruota immagine
        img = cv2.warpAffine(img, rot_mat, (img_size, img_size), borderValue=255)
    
    
    return img
```

```python
def salva_immagini(generator_func, folder, n, prefix):
    for i in range(n):
        img = generator_func()
        filename = os.path.join(folder, f"{prefix}_{i:03d}.png")
        cv2.imwrite(filename, img)



# ------------------------------
# GENERAZIONE TRAINING SET E TEST SET
# ------------------------------

salva_immagini(genera_cerchio,  f"{output_dir}/circles",  num_train, "circle")
salva_immagini(genera_quadrato, f"{output_dir}/squares",  num_train, "square")

salva_immagini(genera_cerchio,  test_dir, num_test, "circle")
salva_immagini(genera_quadrato, test_dir, num_test, "square")

```
